```python
import zipfile
```

```python
# 指定 ZIP 文件的路径
zip_path = 'data/22220_susono-shi_city_2023_citygml_1_op.zip'

# 创建一个 ZipFile 对象
with zipfile.ZipFile(zip_path, 'r') as zfile:
    # 获取 ZIP 文件内所有文件的列表
    file_names = zfile.namelist()
    print("Files in zip:", file_names)
    
    # 假设我们知道要读取的文件名
    folder_to_extract = 'udx/bldg'
    
    # 检查并解压该文件夹中的所有文件
    for file in file_names:
        if file.startswith(folder_to_extract):
            # 解压单个文件
            zfile.extract(file, path='data/susono/')  # 解压到指定目录

    print(f"Extracted contents of {folder_to_extract} to 'data/susono/'")
```
